```python
#non eda
#groupby
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from sklearn.datasets import load_iris
warnings.filterwarnings('ignore')
```

```python
iris = load_iris()
x = iris.data
y = iris.target
df = pd.DataFrame(x, columns=iris.feature_names)
df['target'] = y
df.head()
```

```python
df=pd.read_csv('/content/IRIS.csv')
df.head()
```

```python
#categorical and continous
df.groupby('target') ['petal length (cm)'].agg(['mean','min','count','sum'])
```

```python
#analysing two categorical values
pd.crosstab(df['target'],df['target'])
```

```python
#scatter plots are used whenever we have the continous value)
sns.scatterplot(data=df,x='petal length (cm)', y='petal width (cm)',hue='target')
```

```python
g=sns.FacetGrid(data=df, col='target')
g.map(sns.scatterplot, 'petal length (cm)','petal width (cm)')
```
