```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
%matplotlib inline
```

```python
iris = load_iris()
x = iris.data
y = iris.target
df = pd.DataFrame(x, columns=iris.feature_names)
df['target'] = y
df.head()
```

```python
df.dtypes
```

```python
df.info()
```

```python
#scatter plot
plt.scatter(df.index,df['petal width (cm)'])
plt.show()
```

```python
sns.scatterplot(data=df,x=df.index, y='petal width (cm)',hue=df['target'])
```

```python
#swarm plot
sns.swarmplot(data=df,y='petal width (cm)')
```

```python
sns.swarmplot(data=df,y='petal width (cm)',x='target')
```

```python
#histogram chart also used for continous dataset
plt.hist(df['petal width (cm)'])
```

```python
#density plot
df['petal width (cm)'].plot(kind='density')
```

```python
#box plot
plt.boxplot(df['petal width (cm)'])
```

```python
sns.boxplot(data=df, x ='sepal width (cm)')
```

```python
df['target'].value_counts().plot.bar()
```

```python
sns.countplot(data=df,x='target')
```

```python
plt.pie(df['target'].value_counts(),labels=iris.target_names)
```
