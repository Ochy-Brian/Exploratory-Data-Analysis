```python
#plotting
plt.figure(figsize=(12,8))

plt.subplot(2,2,1)
plt.scatter(df['Age'],df['Gls'])
plt.xlabel('Age')
plt.ylabel('Goals')
plt.title('Age vs Goals')

plt.subplot(2,2,2)
plt.scatter(df['Age'], df['Ast'])
plt.xlabel('Age')
plt.ylabel('Assists')
plt.title('Age vs Assists')


plt.subplot(2,2,3)
plt.scatter(df['Age'], df['G+A90'])
plt.xlabel('Age')
plt.ylabel('G+A per 90')
plt.title('Age vs Goals + Assists per 90')

plt.subplot(2,2,4)
plt.scatter(df['xG90'], df['Gls90'])
plt.xlabel('xG per 90')
plt.ylabel('Goals per 90')
plt.title('xG90 vs Gls90')
plt.tight_layout()
plt.show()
```

```python
#bar chart of the number of players for each squad
plt.figure(figsize=(10,6))
plt.bar(df['Squad'], df['# Pl'])
plt.xticks(rotation=90)
plt.xlabel('Squad')
plt.ylabel("# Players")
plt.title('Number of Players per Squad')
plt.show()
```

```python
#bar chart of the total goals for each squad
plt.figure(figsize=(10,6))
plt.bar(df['Squad'], df['Gls'])
plt.xticks(rotation=90)
plt.xlabel('Squad')
plt.ylabel('Total Goals')
plt.title('Total Goals per Squad')
plt.show()
```

```python
#histogram of player age distributiin across all squads
plt.figure(figsize=(10,6))
plt.hist(df['Age'], bins=10, edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Player Age Distribution')
plt.show()
```

```python
# Box plot of goals per 90 minutes (Gls90) by squad
plt.figure(figsize=(10, 6))
plt.boxplot([df[df['Squad'] == squad]['Gls90'] for squad in df['Squad']])
plt.xticks(range(1, len(df['Squad'].unique()) + 1), df['Squad'].unique(), rotation=90)
plt.xlabel('Squad')
plt.ylabel('Goals per 90 Minutes')
plt.title('Goals per 90 Minutes by Squad')
plt.show()
```

```python
# Stacked bar chart showing the distribution of goals (Gls) and assists (Ast) for each squad
plt.figure(figsize=(10, 6))
plt.bar(df['Squad'], df['Gls'], label='Goals')
plt.bar(df['Squad'], df['Ast'], bottom=df['Gls'], label='Assists')
plt.xticks(rotation=90)
plt.xlabel('Squad')
plt.ylabel('Count')
plt.title('Distribution of Goals and Assists per Squad')
plt.legend()
plt.show()
```

```python
#pie chart showing the percentage distribution of goals among all squads
plt.figure(figsize=(10,6))
plt.pie(df['Gls'], labels=df['Squad'], autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Percentage Distribution of Goals among Squads')
plt.show()
```

```python
# Heatmap showing the correlation matrix between different variables like age, goals, assists, etc.
plt.figure(figsize=(12, 8))
numeric_df = df.select_dtypes(include=['number'])  # Select only numeric columns
corr_matrix = numeric_df.corr()
plt.imshow(corr_matrix, cmap='coolwarm', interpolation='nearest')
plt.colorbar()
plt.xticks(range(len(corr_matrix.columns)), corr_matrix.columns, rotation=90)
plt.yticks(range(len(corr_matrix.columns)), corr_matrix.columns)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
#scatterplot showing the relationship between player age and goals per 90minutes
plt.figure(figsize=(10,6))
sns.scatterplot(x='Age', y='Gls90', data=df)
plt.xlabel('Player Age')
plt.ylabel('Goals per 90 Minutes')
plt.title('Relationship between Player Age and Goals per 90 Minutes')
plt.show()
```
